# Ejercicio 4:

Verificar que los proyectos que tienen nombres duplicados también tienen el hash de nombre duplicado


**Función que asigna a cada valor de la columna nombre, un string hexadecimal siempre que el valor no sea float, de ser float, devuelve un cateo a string** 

```
def hashit(val):
    if isinstance(val, float): 
        return str(val)
    return md5(val.encode('utf-8')).hexdigest()
```

```
duplicados['name'].apply(hashit).head(12)
```

**Observamos que para un mismo nombre se obtiene el mismo hash**

**Función que busca caracteres especiales o raros**

```
def weird(val):
    if isinstance(val, float): 
        return 0
    return bad.sequence_weirdness(val)
```

**Generamos una columna que contiene el valor de salida de la función anterior, es decir, el número de caracteres especiales por nombre o instancias raras en la cadena que le pasamos, para cada dataset**

```
kickstarter_2018_pandas['name_weirdness'] = kickstarter_2018_pandas['name'].apply(weird)
kickstarter_2018_pandas[kickstarter_2018_pandas['name_weirdness'] > 1][['name','name_weirdness']].head(6)
```

```
kickstarter_2018_iso['name_weirdness'] = kickstarter_2018_iso['name'].apply(weird)
kickstarter_2018_iso[kickstarter_2018_iso['name_weirdness'] > 1][['name','name_weirdness']].head(6)
```

**Obtenemos la cantidad de nombres con caracteres raros en cada dataset**

```
print('Con encoding iso-8859 se encuentran ****',len(kickstarter_2018_iso[kickstarter_2018_iso['name_weirdness'] > 1]),'**** nombres con caracteres especiales\n Mientras que sin encoding se tienen solo ****',len(kickstarter_2018_pandas[kickstarter_2018_pandas['name_weirdness'] > 1]),'****')
```
